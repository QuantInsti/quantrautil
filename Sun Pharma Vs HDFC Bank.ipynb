```python
# Download the data from yahoo finance
import yfinance as yf
sunpharma = yf.download('SUNPHARMA.NS', '2016-1-1', '2019-7-31')['Adj Close']
hdfcbank = yf.download('HDFCBANK.NS', '2016-1-1', '2019-7-31')['Adj Close']
```

```python
# Combine the data into single dataframe
import pandas as pd
data = pd.concat([sunpharma,hdfcbank], axis=1)
data.columns = ['SUNPHARMA', 'HDFCBANK']
data.head()
```

```python
# Plot the cumulative returns
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style("whitegrid")
(data.pct_change()+1).cumprod().plot(figsize=(10,7))
plt.ylabel('Cumulative Percentage (%)')
plt.xlabel('Year-Month')
plt.title('Sun Pharma Vs. HDFC Bank')
plt.show()
```
